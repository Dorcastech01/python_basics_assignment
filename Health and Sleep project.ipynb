```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
healthsleep_df = pd.read_csv('Datasets/Health_Sleep_Statistics.csv', sep=',')
```

```python
print(healthsleep_df)
```

```python
#This dataset is used to show the advantages of a good sleep quality in male and female of various age range. The sleep quality range from 1 - 10; 1 which is very poor sleep quality and 10 which is a very good sleep quality
healthsleep_df.info()
```

```python
female_df = healthsleep_df[(healthsleep_df['Age'] >= 18) & (healthsleep_df['Age'] <= 60) & (healthsleep_df['Gender'].isin(['f'])) & (healthsleep_df['Sleep Quality'] >=6)]
```

```python
print(female_df)
```

```python
print(female_df[['Age', 'Gender', 'Sleep Quality']])
female_df.shape[0]
```

```python
male_df = healthsleep_df[(healthsleep_df['Age'] >= 18) & (healthsleep_df['Age'] <= 60) & (healthsleep_df['Gender'].isin(['m'])) & (healthsleep_df['Sleep Quality'] >=6)]
```

```python
print(male_df)
```

```python
print(male_df[['Age', 'Gender', 'Sleep Quality']])
male_df.shape[0]
```

```python

filtered_df = healthsleep_df[(healthsleep_df['Age'] >= 18) & (healthsleep_df['Age'] <= 60) & (healthsleep_df['Sleep Quality'] >= 6) & (healthsleep_df['Sleep Quality'] <= 10)]


male_data = filtered_df[filtered_df['Gender'] == 'm']
female_data = filtered_df[filtered_df['Gender'] == 'f']


plt.figure(figsize=(10, 6))
plt.hist([male_data['Sleep Quality'], female_data['Sleep Quality']], bins=10, label=['Male', 'Female'], color=['blue', 'fuchsia'])
plt.xlabel('Sleep Quality')
plt.ylabel('Number of People')
plt.title('Sleep Quality Comparison (Male vs. Female) - Ages 18-60')
plt.legend()
plt.grid(axis='y')
plt.show()
```
